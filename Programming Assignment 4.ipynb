### 1. Using the given dataset sent, create the following data frames based on the format provided.<br>
Example: Vis = [“Name”, “Gender”, “Track”, “Math<70”]; hometown is constant as <strong>Visayas</strong>

```python
import pandas as pd # panda library
board = pd.read_csv("board2.csv") # read the given dataset
```

#### a. Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

```python
# Filters for the track to be all Instrumentation, the hometown to be all luzon, and Electronics that are higher than 70
Instru = board[(board["Track"] == "Instrumentation") 
        & (board["Hometown"] == "Luzon") 
        & (board["Electronics"] > 70)]
Instru = Instru[["Name", "GEAS", "Electronics"]] # Only shows the Name, GEAS and, Electronics.
Instru
```

#### b. Filename: Mindy = ["Name", "Track", "Electronics", "Average >=55"]; where hometown is constant as Mindanao and gender Female

```python
board['Average'] = board[['Math', 'Electronics', 'GEAS', 'Communication']].mean(axis=1) # adds a new column for average
# Filters for hometown to be only Mindanao and the gender only to be Female. It also only takes the average that is equal or higher than 55.
Mindy = board[(board["Hometown"] == "Mindanao")
        & (board["Gender"] == "Female")
        & (board["Average"] >= 55)]
Mindy = Mindy[["Name", "Track", "Electronics", "Average"]] # Only shows the Name, Track, Electronics, and Average.
Mindy
```

#### 2. Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for average grade by Track
plt.figure()
sns.boxplot(x="Track", y="Average", data=board)
plt.title("Average Grade by Track")

# Boxplot for average grade by Gender
plt.figure()
sns.boxplot(x="Gender", y="Average", data=board)
plt.title("Average Grade by Gender")

# Boxplot for average grade by Hometown
plt.figure()
sns.boxplot(x="Hometown", y="Average", data=board)
plt.title("Average Grade by Hometown")

```
